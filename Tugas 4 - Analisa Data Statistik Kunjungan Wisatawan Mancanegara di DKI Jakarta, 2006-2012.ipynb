## 3. Visualisasi Sederhana

```python
ax = stat_wisata_dki[stat_wisata_dki['indikator_wisata']=='Jumlah Wisman']
ax = ax.groupby(['tahun']).sum()
# ax = stat_wisata_dki.groupby(['tahun','indikator_wisata']).sum()
ax = ax.plot(kind='line', title ="Gambar 1.1 Statistik Jumlah Kunjungan Wisatawan DKI Jakarta 2006-2012 (Juta Kunjungan)", figsize =(10,10), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
bx = stat_wisata_dki[stat_wisata_dki['indikator_wisata']!='Jumlah Wisman']
bx = bx.groupby(['tahun','indikator_wisata']).sum()
#print(bx)
bx = bx.plot.barh(stacked=True, title ="Gambar 1.2 Statistik Jumlah Kunjungan Wisatawan DKI Jakarta 2006-2012 Berdasarkan Kelompok Hotel (Persentase)", figsize =(10,10), legend=True, fontsize=12)
bx.set_xlabel("Tahun", fontsize=12)
bx.set_ylabel("Jumlah", fontsize=12)
```

## 4. Narasi Sederhana 

Gambar 1.1 menunjukkan Statistik Jumlah Kunjungan Wisatawan Mancanegara (Wisman) di provinsi DKI Jakarta dari tahun 2006 sampai tahun 2012 dalam juta kunjungan. Jumlah kunjungan tersebut ditotal berdasarkan Tingkat Penghunian Kamar Hotel Berbintang dan Kamar Hotel NonBintang (Gambar 1.2). 

Pada tahun 2006, jumlah kunjungan Wisman di provinsi DKI Jakarta adalah sebanyak 1.22 juta kunjungan. Dari 1.22 juta kunjungan tersebut, sebanyak 50.60 persen tamu menginap di Hotel Bintang dan 57.34 persen tamu menginap di Hotel Non-Bintang. 

Kemudian Pada tahun berikutnya, tidak terjadi peningkatan kunjungan Wisman. Namun demikian, terjadi penurunan persentase tamu yang menginap di Hotel-Bintang sebesar minus 0.15 persen dibanding tahun sebelumnya. 

Tahun 2008, terjadi peningkatan kunjungan Wisman yang cukup signifikan. Jumlah tersebut meningkat sebanyak 0.33 juta kunjungan atau 27.05 persen dibanding tahun sebelumnya. Namun hal tersebut tidak bertahan lama. Tahun berikutnya, jumlah kunjungan wisman justru mengalami penurunan sebesar 0.08 juta kunjungan. 

Pada tahun 2010, jumlah wisman kembali meningkat. Dari Jumlah kunjungan sebanyak 1.45 juta kunjungan pada tahun 2009 naik sebesar 0.44 menjadi 1.89 juta kunjungan. 

Setelah terjadi fluktuasi selama tahun 2008-2010, terjadi peningkatan kunjungan wisman yang cukup stabil dari tahun 2011-2012. Pada tahun 2011, untuk pertama kalinya sejak 2006 jumlah kunjungan wismaan mencapai 2 juta kunjungan. Kemudian pada tahun 2012, terjadi lagi peningkatan kunjunga sebesar 6.5 persen dibanding tahun 2011, menjadi 2.13 juta kunjungan
